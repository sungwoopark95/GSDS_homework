```python
# 크롤링 코드!
user_info=[]
# ----------아래 watcha_movie_id[] 안 인덱싱 변경!!!!!------------
# 200:450 - 세라님, 450:700,- 상수님 700:950 - 동근 950:1200 - 성우님, 1200: - 지상님(1200~1410)
for movie_id in watcha_movie_id[시작인덱스:끝인덱스]:
    print("movie id:",movie_id)
    url = f"https://pedia.watcha.com/en-KR/contents/{movie_id}/comments"
    wait = WebDriverWait(driver,10)
    driver.get(url)
    time.sleep(7)
    try:
        driver.find_element_by_class_name("css-1d6nfxu").click() # 웹툰 팝업창 닫기
    except:
        pass
    for item in range(3): # 댓글 스크롤
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(4)

    # user,rating 저장 리스트
    users=[(user.find_element(by=By.CSS_SELECTOR, value='a').get_attribute("href").split('/')[-1],user.text.split()[-1]) for user in driver.find_elements_by_class_name("css-4obf01")]
    print("User number:",len(users))
    time.sleep(5)
    for user,rating in users:
        dic={}
        print("user: ",user)
        dic['watcha_movie_id']=movie_id
        dic['user_id']=user
        dic['rating']=rating
        #print("movie:",movie_id)
        user_url = f"https://pedia.watcha.com/en-KR/users/{user}/analysis"
        driver.get(user_url)
        #print("user page enter")
        time.sleep(10)
        try:
            user_page=driver.find_elements_by_class_name("css-1f4aos9")
            time.sleep(3)
            for user_block in user_page:
                user_page_text=user_block.text
                if "Favorite Actors" in user_page_text:
                    dic["Favorite_Actors"]=user_page_text.split('\n')[1:]
                elif "Favorite Directors" in user_page_text:
                    dic["Favorite_Directors"]=user_page_text.split('\n')[1:]
                elif "Favorite Movie Genres" in user_page_text:
                    dic["Favorite_Genres"]=user_page_text.split('\n')[1:]
                #print("error ㅇㄷ")
        except:
            print("error")
            pass
        user_info.append(dic)
```
